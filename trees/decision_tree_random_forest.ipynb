```python
import pandas as pd, numpy as np
names=['age', 'job', 'marital', 'educatin', 'default', 'balance', 'housing', 'loan', 'contact', 'day', 'month', 'duration', 'campaign', 'pdays', 'previous', 'poutcome', 'y']
df = pd.read_csv('bank-full.csv',delimiter=';', header=0 )
```

```python
df.info()
```

```python
df = df.replace(['yes', 'no'], [1, 0])
df.head(3)
```

```python
df.info()
```

```python
df['default'].describe()
```

```python
# Check for any null values
df.isnull().values.any()
```

```python
df = pd.get_dummies(df, columns=['job', 'marital', 'education', 'contact', 'month', 'poutcome'])
df.head(3)
```

```python
df.info()
```

```python
xdf = df[['age', 'default', 'balance', 'housing', 'loan', 'day', 'duration', 'campaign', 'pdays', 'previous', 'job_admin.', 'job_blue-collar', 'job_entrepreneur', 'job_housemaid', 'job_management', 'job_retired', 'job_self-employed', 'job_services', 'job_student', 'job_technician', 'job_unemployed', 'job_unknown', 'marital_divorced', 'marital_married', 'marital_single', 'education_primary', 'education_secondary', 'education_tertiary', 'education_unknown', 'contact_cellular', 'contact_telephone', 'contact_unknown', 'month_apr', 'month_aug', 'month_dec', 'month_feb', 'month_jan', 'month_jul', 'month_jun', 'month_mar', 'month_may', 'month_nov', 'month_oct', 'month_sep', 'poutcome_failure', 'poutcome_other', 'poutcome_success', 'poutcome_unknown']]
ydf = df[['y']]
x = np.asarray(xdf)
y = np.array(ydf)
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
unique, counts = np.unique(y_train, return_counts=True)
print("y_train")
print(np.asarray((unique, counts)).T)

print("")

unique, counts = np.unique(y_test, return_counts=True)
print("y_test")
print(np.asarray((unique, counts)).T)
```

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier(random_state=0)
clf = clf.fit(x_train, y_train)
```

```python
from sklearn import metrics
from pandas_ml import ConfusionMatrix

predictions = clf.predict(x_test)
predict_proba = clf.predict_proba(x_test)

prec = metrics.precision_score(y_test, predictions)
rec = metrics.recall_score(y_test, predictions)
f1 = metrics.f1_score(y_test, predictions)   
#logloss=metrics.log_loss(y_test, predict_proba)
mse = metrics.mean_squared_error(y_test, predictions)
accuracy = metrics.accuracy_score(y_test, predictions)
prfs = metrics.precision_recall_fscore_support(y_test, predictions)

predictions_train = clf.predict(x_train)
predict_proba_train = clf.predict_proba(x_train)

prec_tr = metrics.precision_score(y_train, predictions_train)
rec_tr = metrics.recall_score(y_train, predictions_train)
f1_tr = metrics.f1_score(y_train, predictions_train)   
mse_tr = metrics.mean_squared_error(y_train, predictions_train)
accuracy_tr = metrics.accuracy_score(y_train, predictions_train)
prfs_tr = metrics.precision_recall_fscore_support(y_train, predictions_train)

print("precision_test: ", prec, "\nprecision_train: ", prec_tr)
print("\n\nrecall_test: ", rec, "\nrecall_train: ", rec_tr)
print("\n\nf1_test: ", f1, "\nf1_train: ", f1_tr)
print("\n\nmse_test: ", mse, "\nmse_train: ", mse_tr)
print("\n\naccuracy_test:", accuracy, "\naccuracy_train", accuracy_tr)
print("\nConfusion Matrix: Test\n")
# print metrics.confusion_matrix(y_test, predictions, [0, 1])
cm = ConfusionMatrix(list(map(lambda a: a[0],y_test)), predictions)
print(cm)

print("\nConfusion Matrix: Train\n")
#print metrics.confusion_matrix(y_test, predictions, [0, 1])
cm_tr = ConfusionMatrix(list(map(lambda a: a[0],y_train)), predictions_train)
print(cm_tr)
```

```python
from sklearn.model_selection import GridSearchCV

parameters= { "max_depth":[3, 5, 7, 9], "min_samples_leaf":[ 1, 2, 4], "min_samples_split":[2, 4, 6]}   
gridSearch = GridSearchCV(DecisionTreeClassifier(random_state=0), param_grid=parameters, cv=5)

c, r = y_train.shape
y_train.reshape(c,)
print(x_train.shape, y_train.shape)
gridSearch.fit(x_train, y_train.reshape(c,))
print("Grid Search: Best score: ", gridSearch.best_score_)  
print("Best Params: ",  gridSearch.best_params_)

selectedParams = gridSearch.best_params_ 
selectedModel = DecisionTreeClassifier(**selectedParams)
selectedModel = selectedModel.fit(x_train, y_train)
```
